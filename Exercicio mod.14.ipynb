```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

### Mês de Março

```python
sinasc_MAR = pd.read_csv('./input/SINASC_RO_2019_MAR.csv')
sinasc_MAR.head()
```

```python
sinasc_MAR.DTNASC.value_counts().sort_values()
```

```python
pd.pivot_table(sinasc_MAR, values='IDADEPAI', index='DTNASC',
               aggfunc='mean').plot(figsize=[17, 7])
plt.ylabel('idade media pai');


pd.pivot_table(sinasc_MAR, values='GESTACAO', index=['ESCMAE'],
                       aggfunc='count').plot(figsize=[18, 6]);


pd.pivot_table(sinasc_MAR, values='IDADEMAE', index=['DTNASC', 'SEXO'],
               aggfunc='count').unstack().plot(figsize=[18, 6]);


pd.pivot_table(sinasc_MAR, values='APGAR5', index=['GESTACAO'],
                   aggfunc='mean').sort_values('APGAR5').plot(figsize=[18, 6])
plt.ylabel('apgar5 médio');
plt.xlabel('gestcao');


pd.pivot_table(sinasc_MAR, values='APGAR1', index=['GESTACAO'],
                   aggfunc='mean').sort_values('APGAR1').plot(figsize=[18, 6])
plt.ylabel('apgar1 médio');
plt.xlabel('gestcao');


pd.pivot_table(sinasc_MAR, values='PESO', index=['APGAR5'],
                   aggfunc='mean').plot(figsize=[18, 6])
plt.ylabel('peso');
plt.xlabel('apgar5 medio');
```
